```
X=[[151],[174],[138],[186],[128],[136],[179],[163],[152],[131]] #2d-format
y=[63,81,56,91,47,57,76,72,62,48] #1d format
```

```
from sklearn.linear_model import LinearRegression
#import ML algo
```

```
regressor=LinearRegression()
#create object
```

```
#train the algorithm
regressor.fit(X,y)
```

```
regressor.predict([[175]])
```

```
regressor.score(X,y)*100  #check accuracy
```

```
import matplotlib.pyplot as plt
```

```
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(X,y)
plt.plot(X,regressor.predict(X),color="red",label="Predicted")
plt.legend()
```
