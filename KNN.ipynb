```
import pandas as pd
import seaborn as snn
```

```
df=pd.read_csv('ingredients.csv')
```

```
df
```

```
snn.countplot(df['class'])
```

```
X=df[['sweetness','crunchiness']]
y=df['class']
```

```
from sklearn.neighbors import KNeighborsClassifier
```

```
clf=KNeighborsClassifier(n_neighbors=4)
```

```
clf.fit(X,y)
```

```
clf.score(X,y)*100
```

```
new=[[6,4]]
```

```
clf.predict(new)
```
