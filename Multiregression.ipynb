```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('mtcars.csv')
```

```
df.shape
```

```
df.columns
```

```
X=df[['disp','hp','wt']]
y=df['mpg']
```

```
from sklearn.linear_model import LinearRegression
```

```
regressor=LinearRegression()
```

```
regressor.fit(X,y)
```

```
regressor.predict([[120,87,2.1]])
```

```
regressor.score(X,y)*100
```

```
  import matplotlib.pyplot as plt
```

```
plt.xlabel('DISP,HP,WT')
plt.ylabel('Mileage per galloon')
plt.plot(X,y)
plt.plot(X,regressor.predict(X),color="black",label="Predicted")
plt.legend()
```
